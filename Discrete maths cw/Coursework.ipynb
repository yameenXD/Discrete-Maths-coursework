```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#Reading in the csv file into my jupyter file and will only read in the age and total column
data = pd.read_csv('4214293_dataset.csv', usecols = ['age', 'total'])
#This will output the columns of age and total
print(data)
age = np.array(data.age)
total = np.array(data.total)
#Creates the scatter graph between age and total
plt.scatter(age, total)
#The title for the graph
plt.title("Age vs.  Total")
#The label for the x-axis
plt.xlabel('Age', fontsize=13)
#The label for the y-axis
plt.ylabel('Total', fontsize=13)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Reading in the csv file
data = pd.read_csv('4214293_dataset.csv', usecols=['status'])
status_counts = data['status'].value_counts()

#Plotting the bar chart
status_counts.plot(kind='bar', color='red')
plt.title("Status of orders")
plt.xlabel('Status', fontsize=13)
plt.ylabel('Frequency', fontsize=13)
plt.show()
```

```python
#Importing libabries 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#Reading in the csv file
data = pd.read_csv('4214293_dataset.csv')
qty_ordered = np.array(data.qty_ordered)
#Title for the graph
plt.title("Quanitity ordered vs Frequency")
#X-axis label
plt.xlabel('Quantity ordered', fontsize=13)
#Y-axis label
plt.ylabel('Frequency', fontsize=13)
#Plotting the histogram which will have a colour brown
plt.hist(qty_ordered, color = 'brown', bins=10)
plt.show()
```

```python
#Pie chart for regions
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#Reading the csv file
data = pd.read_csv('4214293_dataset.csv')
#Plotting the pie chart
region = np.array(data.Region)
region_counts = data['Region'].value_counts()
plt.pie(region_counts, labels = region_counts.index, autopct ='%1.1f%%')
plt.title('Distribution of regions')

plt.show()
```

```python
#Barchart for different states
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#Reading in the csv file
data = pd.read_csv('4214293_dataset.csv')
#Plotting the bar chart
state_counts = data['State'].value_counts()
state_counts.plot(kind='bar', color = 'purple', width=1)

#Title of the bar chart
plt.title('Different states')
#The x-axis label
plt.xlabel("States", fontsize=13)
#The y-axis label
plt.ylabel("Frequency", fontsize=13)
plt.xticks(rotation=90, ha='right')
plt.tight_layout()


plt.show()
```
